```python
# To help you get started...
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 


data = pd.read_csv('~/Desktop/exampleco_data/machine_1.csv',index_col=0)
data.reset_index(inplace = True)

plt.plot(range(len(data)), data[["0", "1", "2", "3"]])
plt.show()
```

```python
#I explored several different ways of removing the noise from the data. 
#1. Rolling average - I decided against this because there was still noise after applying a rolling average and it would not capture the change points as well
#2. Clipping - I decided against this as well because I didn't want to manipulate data points when I am tring to find trends
#3. Remove - I chose this method. I determine that about 5% of the data was noise so I removed that 5%
    #I removed by setting outliers to None. I wanted to do this to ensure that the data remained aligned with the timestamps

#the second step was to find the faulty and failure stages. 
#I noticed in plots of the time series data that the data oscillated to a greater degree when in the faulty stage. 
#This oscillation suggested to me that maybe the standard deviation was the best metric to measure when the faulty stage occurs.
#So I took a rolling standard deviation (10 time stamps at a time)
#This may seem somewhat arbitrary, but data science is inherently an imperfect science. 
#I've learned that sometimes there isn't a statistical justification for every decision. 
    
#Function takes in the data and returns a df with the rolling stdevs for each series in a new df
def get_rolling(df):
    rolling_df = pd.DataFrame()
    rolling_df["Time"] = df["index"]

    for i in range(4):
        #copy of data for manipulation
        current = df[str(i)].copy()
        #identify <2.5% and >97.5% 
        lower = current.quantile(.03)
        upper = current.quantile(.97)
        #Set outliers to None
        current.iloc[np.where(current < lower)] = None
        current.iloc[np.where(current > upper)] = None
        #assign new data to new dataframe    
        #take rolling std
        rolling_df[str(i)] = pd.Series(current).rolling(10).std()
        
    return rolling_df

#This is a separate function that just returns a df with the noise removed. This is easier for plotting 
def remove_noise(df):
    no_noise = pd.DataFrame()
    no_noise["Time"] = df["index"]

    for i in range(4):
        #copy of data for manipulation
        current = df[str(i)].copy()
        #identify <2.5% and >97.5% 
        lower = current.quantile(.03)
        upper = current.quantile(.97)
        #Set outliers to None
        current.iloc[np.where(current < lower)] = None
        current.iloc[np.where(current > upper)] = None
        no_noise[str(i)] = current

    return no_noise
```

```python

#Here is the methodology for finding the fault and fail points. 
#The standard deviation increases substantially at the fault points. 
#Therefore, once the standard deviation is above the 98th percentile for 3 timesteps, a fault has occurred.
#In fail mode, I was able to simply pinpoint when the values were very lose to zero
#. Therefore, once the no noise value is betweeen -0.05 and 0.05, a fail has occurred
#I returned the mean of the fault/fail points across the 4 sets of time series data.
#Therefore, if there is an outlier or one of the timeseries does not produce a fault/fail, there is still a
#reasonable estimate. 

#Function takes in a dataframe of the rolling stdev time series data, and returns this fault point - 10
def find_fault(df):
    fault_points = []
    for i in range(3):
        #set fault threshold to be 98% of stdev for the series
        thresh = df[str(i)].quantile(.98)
        #set fail threshold to be 25% of stdev for the series
        low_thresh = df[str(i)].quantile(.25)
        
        for n in range(2,len(df)):
            #select window of points
            prev2 = df[str(i)][n-2]
            prev1 = df[str(i)][n-1]
            curr = df[str(i)][n]
            
            #if all three points are above the threshold, call it a fault point, break the loop
            if curr > thresh and prev1 > thresh and prev2 > thresh:
                fault_points.append(n)
                break
                
            #if we reach a failure, break the loop
            elif curr < low_thresh and prev1 < low_thresh and prev2 < low_thresh:
                break
    
    return np.median(fault_points)

#Function takes in a dataframe of the no noise data, returns the calculated fail point
#fail point
def find_fail(df):
    fail_points = []
    for i in range(3):
        #set threshold to be 0.05 and -0.05, so only values very close to zero are captured
        high_thresh = 0.05
        low_thresh = -0.05
        for n in range(2,len(df)):
            #select window of points
            prev2 = df[str(i)][n-2]
            prev1 = df[str(i)][n-1]
            curr = df[str(i)][n]
            
            #if all three points are below/above the fail threshold, call it a fail and break the loop
            if curr < high_thresh and prev1 < high_thresh and prev2 < high_thresh \
            and curr > low_thresh and prev1 > low_thresh and prev2 > low_thresh:
                fail_points.append(n)
                break

    return np.median(fail_points)
    

        
        
```

```python
#here is a code block that takes in a given data file, gets the rolling frame, the fault, and the failure,
#and then plots the fault and failure lines on a graph with the data (noise removed)

#if you change the directory and the file number, you can try this out for each file provided (0-19)
data = pd.read_csv('~/Desktop/exampleco_data/machine_1.csv', index_col=0)
data.reset_index(inplace = True)
rolling = get_rolling(data)
fault = find_fault(rolling)
data_no_noise = remove_noise(data)
fail = find_fail(data_no_noise)

if fault < fail:
    print("Fault TimeStep: ", (fault))
else: 
    print("Fault TimeStep: ", "nan")
print("Fail TimeStep: ", (fail))

plt.plot(range(len(data)), data_no_noise[["0", "1", "2", "3"]])
plt.vlines(fault,-100, 100, colors = 'k', label = "FAULT")
plt.vlines(fail,-100, 100, colors = 'k', label = "FAIL")
plt.xlabel("Time Step #")
plt.ylabel("Data Value")
plt.title("Time Series Data with Fault/Fail Lines")
plt.legend(["0", "1", "2", "3"])
plt.show()
```
